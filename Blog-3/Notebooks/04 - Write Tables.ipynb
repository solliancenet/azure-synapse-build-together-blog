```python
from azureml.opendatasets import NycTlcGreen
import datetime

data = NycTlcGreen(datetime.datetime(2014, 1, 1, 0, 0), datetime.datetime(2014, 12, 31, 0, 0))
df = data.to_spark_dataframe()
```

```python
df.createOrReplaceTempView("2014TaxiData")
```

```python
%%sql
SELECT 
    COUNT(*)
FROM 
    2014TaxiData
```

```python
spark.sql("CREATE DATABASE IF NOT EXISTS nyctaxi")

df.write.mode("overwrite").saveAsTable("nyctaxi.2014TaxiData")
```

```python
df = spark.sql("SELECT * FROM nyctaxi.2014TaxiData")
df.show(10)
```
